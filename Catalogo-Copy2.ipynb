```python
from astroquery.utils.tap import TapPlus

# Connect to the NASA Exoplanet Archive TAP service
tap_service = TapPlus(url="https://exoplanetarchive.ipac.caltech.edu/TAP")

# Define the query to retrieve the host star names and effective temperatures
query = "SELECT pl_hostname, st_teff FROM ps"

# Submit the query and retrieve the results as a Pandas DataFrame
results = tap_service.launch_query(query).to_table().to_pandas()

# Print the results
print(results)
```
