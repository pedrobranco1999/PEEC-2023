```python
#If we ran vizier fro every star at the same time (not divided in groups) then the 'repitions' are the same 
#for each catalog and already refer to the global count

catalogs_df = pd.read_csv('catalogs.txt', usecols=[1,2])

# eliminating the lines that have the same catalog (the number is also the same because we counted everything already)
catalogs_df = catalogs_df.drop_duplicates()


better_cat = []
nums = []
for name in catalogs_df['catalogs']:
    parts = name.split('/')
    new = parts[:-1]
    new = [i+'/' for i in new]
    new = ''.join(new)
    num = float(catalogs_df[catalogs_df['catalogs']==name]['repetitions'])
    if new not in better_cat:
        better_cat.append(new)
        nums.append(num)
    else:
        for j,n in enumerate(better_cat):
            if n == new:
                nums[j] += num
# print(better_cat)
# print(nums)
# print(len(better_cat))
# print(len(nums))

#making it into a dataframe
dic = {'big_catalogs': better_cat, 'number_of_appearances': nums}
reduced_df = pd.DataFrame(dic)
```
